# What are embeddings?

Word embeddings are central in language model (LLM) applications. They are essentially dense vector representations of words that capture semantic relationships. 

Embeddings are trained through unsupervised learning on large text corpora, where words with similar meanings are positioned close to each other in the vector space. 

Word embeddings enhance the model's understanding of language by providing knowledge about relationships between words, enabling more accurate predictions and context-aware responses. These embeddings enable a degree of semantic understanding.

```python
import os
import seaborn as sns
from sklearn.metrics.pairwise import cosine_similarity # cosine similarity to compare embeddings

from dotenv import load_dotenv # for loading environment variables
from langchain_openai import OpenAIEmbeddings, AzureOpenAIEmbeddings
```

Here, let's use embeddings from OpenAI:

```python
def get_embedder(embedder_name: str) -> AzureOpenAIEmbeddings:
    """Obtain an instance of the OpenAIEmbeddings class - a wrapper around the OpenAI API embedding functionality.

    Args:
        embedder_name (str): type of embedder to use

    Raises:
        ValueError: if the embedder_name is not a valid embedder name

    Returns:
        OpenAIEmbeddings: an instance of the OpenAIEmbeddings class
    """
    if embedder_name == "text-embedding-ada-002":
        return OpenAIEmbeddings(
            model=embedder_name,  # just the model name
            openai_api_key=os.getenv("OPENAI_API_KEY"),
        )
    else:
        raise ValueError(f"{embedder_name} is not a valid embedder name")
```

A classic embedding space is ada:

```python
embedder = get_embedder("text-embedding-ada-002")
```

```python
embeddings = embedder.embed_query("This is a test query")

ax = sns.lineplot(x=range(len(embeddings)), y=embeddings)
ax.set(ylabel='Embedding Value', xlabel='Dimension', title='Embeddings')
```

So we can now embed text. We can also use this to assess the similarity between text:

```python
embedding_1 = embedder.embed_query("This is a test query")
embedding_2 = embedder.embed_query("This is another test query")

similarity_score = cosine_similarity([embedding_1], [embedding_2])[0, 0]

print(f"The similarity score is {similarity_score}")
```

```python
# Here are a few other embeddings:

from sklearn.metrics.pairwise import euclidean_distances
from sklearn.metrics.pairwise import manhattan_distances

# Eucledian distance is the square root of the sum of squared differences between corresponding elements of the two vectors.
# Useful when the magnitude of vectors is important.
euclidean_distance = euclidean_distances([embedding_1], [embedding_2])[0,0]
print(f"The Euclidean distance is {euclidean_distance}")

# Manhattan distance is the sum of absolute differences between corresponding elements of the two vectors.
# Useful when dealing with data with different scales or sparse data.
manhattan_distance = manhattan_distances([embedding_1], [embedding_2])[0, 0]
print(f"The Manhattan distance is {manhattan_distance}")
```

Create a database:

```python
# Let's simulate a database of embeddings>
database_text = [
    "This is a test query",
    "This is another test query",
    "We are building a chatbot in a day",
    "It's raining cats and dogs",
    "I am about to make some tea, would you like some?",
]

# Let's embed the database text>
database_embeddings = [embedder.embed_query(text) for text in database_text]
```

Create a query and find the best matching text in the database:

```python
query = "This is a test"
# query = "I'm hungry"
# query = "What's really missing in my life is a pet"

embedded_query = embedder.embed_query(query)

similarity_score = cosine_similarity([embedded_query], database_embeddings)[0].tolist()

most_similar_text = database_text[similarity_score.index(max(similarity_score))]

print(f'The best matching text from the database for your query "{query}" is "{most_similar_text}".')
```
