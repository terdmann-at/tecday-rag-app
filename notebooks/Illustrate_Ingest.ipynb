# What is data ingestion?

During data ingestion, we use langchain functionalities to read documents, represent them as specified Langchain Document Classes and save them into our database

Here's a simple example of the process of reading in the documents and defining the document class info in ingest.py and model.py (Tasks 2.1-2.4):

```python
from langchain.document_loaders.pdf import PyMuPDFLoader 
```

```python
pdf_loader = PyMuPDFLoader(str('data/raw/FAQ Working conditions.pdf'))

pdf_loader
```

```python
pdf_pages = pdf_loader.load_and_split()

print(f'Content: \n {pdf_pages[0].page_content}')
print('##########')
print(f'Source: \n {pdf_pages[0].metadata["source"]}')
print('##########')
print(f'Page: \n {pdf_pages[0].metadata["page"]}')
```
